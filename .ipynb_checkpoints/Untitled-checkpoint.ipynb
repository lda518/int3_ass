```python
%matplotlib inline

# Store the best models of each experiment here
best_models = {}
# TODO normalize data from the beginning with standardScaler??????

# Import data ignoring first row since we will not be making any calculations with these labels
data = np.genfromtxt('data.csv',delimiter=',')[1:,:]
data = StandardScaler().fit_transform(data)
# Extract inputs, ignoring the last row based on the constraints of the assignment
X = data[:-1,:] 
# Take D as the target, one time lag down from the inputs
y = data[1:,-1] 
```

```python
# TODO eventually look into intercept properly, looks like = False gives better results?
# TODO normalize data!
training_sizes = []
for i in range(20,len(X)-20, 5):
    training_sizes.append(i)

def eval_model(models, X=X):
    # Keep track of the r2 scores and the corresponding models for final evaluation
    model_scores = {}
    r2_scores = []
    # Go through different sizes of training data to find a balance between overfitting and
    # underfitting
    for train in training_sizes:
        X_train = X[:train]
        X_test = X[train:]
        y_train = y[:train]
        y_test = y[train:]
        # Cylce through and fit the models given for evaluation
        for model in models:
            if model == LinearRegression:
                reg = model()
            else:
                reg = model()
            reg.fit(X_train, y_train)

            # Extract the scores for each model and associate with the model in a dictionary
            y_pred = reg.predict(X_test)
            score = r2_score(y_test, y_pred)
            #print(score)
            r2_scores.append(score)
            model_scores[score] = [reg, train]
            #adj_r2 = 1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
    optim_model, score = get_best(r2_scores, model_scores)
    return optim_model, score 
```

```python
def get_best(r2_scores, model_scores):
    best = np.max(r2_scores)
    print('The best performing model in the run was: {}\n\
          With an r2 score of: {}\n\
          With a training data size of: {}\n'
          .format(model_scores[best][0], best, model_scores[best][1]))
    return model_scores[best][0], best
```

```python
def graph(model):
    y_pred = model.predict(X)
    x_axis = []
    for i in range(y.size):
        x_axis.append(i)
    plt.figure(figsize=(10, 6))
    plt.title("Fitted data", size=16)
    plt.scatter(x_axis, y)
    plt.plot(x_axis, y_pred, c="red")
    plt.show()
```

Evaluation of models and model selection
The primary features we want out of the model are r2 accuracy, and ability to generalize to other training sets
Becuase of this, we are aiming to maximize the r2 value while trying to minimize the bias learned in them model
All of the previously observed models have comparable r2 scores. 
Since regularisation causes bias, and the r2 score for the basic LinearRegression() model is better, I have discarded the models utilizing regularisation
The models utilizing polynomial features have a worse r2 score as well as being more computationally demanding. This suggests that adding polynomial non linearities to the inputs provides no benefit and makes the model worse. For this reason, the polynomial model has been discarded. In fact, the polynomial model has a better r2 score when its degree is set to 1, suggesting that it is at its most effective when simply performing linear regression
Finally, the piecewise regression model has a worse r2 score than the LinearRegression model when at a reasonable degree. If the degree is set to something extreme such as 10, it very marginally outperforms the LinearRegression model. Due to the added complexity of the model computationally and in terms of its features, it suggests overfitting.
The initial LinearRegression model is the one that will be chosen

```python
# Create a daset object for the data so that I can later pass it to a dataloader for training
class LightSourceDataset(Dataset):
    
    def __init__(self, img_dir, labels_file=None, transform=None):
        super(LightSourceDataset, self).__init__()
        self.labels = pd.read_csv(labels_file, header=None)
        self.img_dir = img_dir
        self.transform = transform
    
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.labels.iloc[idx,0])
        # Normalize image data in between -1 and 1 to for better training
        image = (read_image(img_path).float()/255)*2 -1
        #image = (self.data[idx,:,:].float()/255).unsqueeze(0)

        # Reduce resolution of image to reduce unnecessary training for network
        if self.transform:
            image = self.transform(image)
        #resize(img=image, size=30)
        
        label = torch.from_numpy(self.labels.iloc[idx,1:].to_numpy(dtype=float))
        return image, label
```

```python
# Choose a regression model instead of a classification model since we do not know whether
# the evaluation test data will also contain a set of 64 classes
#
# Make the network narrow and deep rather than broad and shallow in order to learn the 
# features better
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.convlayers = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=6, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(6),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(in_channels=6, out_channels=12, kernel_size=3, stride=1, padding=1),
            nn.BatchNorm2d(12),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(in_channels=12, out_channels=24, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(24),
            nn.ReLU(),
            nn.Conv2d(in_channels=24, out_channels=48, kernel_size=3, stride=1, padding=0),
            nn.BatchNorm2d(48),
            nn.ReLU()
        )
        
        self.MLP = nn.Sequential(
            # 24 x 46 x 38 for un-reduced images
            # 24 x 
            nn.Linear(in_features=48*5*3, out_features=500),
            nn.BatchNorm1d(500),
            nn.ReLU(),
            nn.Linear(in_features=500, out_features=200),
            nn.BatchNorm1d(200),
            nn.ReLU(),
            nn.Linear(in_features=200, out_features=3),
        )

    def forward(self,x):
        print(x.shape)
        x = self.convlayers(x)
        print(x.shape)
        
        x = x.view(x.size(0), -1)
        #print('flattened')
        print(x.shape)
        x = self.MLP(x)
        #x = x / torch.linalg.norm(x)
        return x


model = CNN()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Using the absolute error for loss avoids large spikes in accuracy during training, as opposed to with MSE loss
loss_func = nn.L1Loss()
```

```python
images, labels = next(iter(train_loader))
print(images[0].size())
output = model(images)
print(output.shape)
loss = loss_func(output,labels)
print(loss)
```

```python
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(total_params)
```

```python
def unit_vector(vector):
    return vector / np.linalg.norm(vector)
def get_angle(vec1, vec2):
    v1_u = unit_vector(vec1)
    v2_u = unit_vector(vec2)
    return np.rad2deg(np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0)))
```

```python
num_epochs = 200

# Use Adam optimiser since it is the least sensitive to hyperparameters and most likely to
# work robustly
optim_cnn = torch.optim.Adam(model.parameters(), lr = 0.003)

# Initialise some variables to track mean angular errors for visualization
iterations_per_epoch = math.ceil(len(train_data)/batch_size)
training_ang_errors = []
testing_ang_errors = []

model.train()
for epoch in range(num_epochs):
    ang_errors = []
    total_loss = 0
    for i, (images, labels) in enumerate(train_loader):
        images, labels = images.to(device), labels.to(device)
        output = model(images)
        #print(type(output))
        #print(output)
        for j in range(len(output)):
            error = get_angle(output[j].detach().numpy(), labels[j].detach().numpy())
            ang_errors.append(error)
        #print(output.shape)
        #print(labels.shape)
        loss = loss_func(output.float(),labels.float())
        optim_cnn.zero_grad()
        loss.backward()
        optim_cnn.step()
        #print(loss.item())
        if (i+1) % 10 == 0:
           print('Epoch [{}/{}], Iteration [{}/{}], Loss: {:.4f}'.format(epoch + 1, num_epochs, i + 1, iterations_per_epoch, loss.item()))
        total_loss += loss
    sum_ang = sum(ang_errors)
    mean_ang = sum_ang/len(ang_errors)
    training_ang_errors.append(mean_ang)
    print('Mean angular error over training epoch {}: {}'.format(epoch+1, sum_ang/len(ang_errors)))
    #print('Total loss over epoch {}: {:.2f}'.format(epoch+1,total_loss))

    model.eval()
    with torch.no_grad():
        ang_errors = []
        for images, labels in test_loader:
            output = model(images)
            for j in range(len(output)):
                error = get_angle(output[j].detach().numpy(), labels[j].detach().numpy())
                ang_errors.append(error)
        sum_ang = sum(ang_errors)
        mean_ang = sum_ang/len(ang_errors)
        testing_ang_errors.append(mean_ang)

        print('Mean angular error over validation epoch: {}'.format(sum_ang/len(ang_errors)))
    model.train()
```

```python
plt.title("Training curve")
plt.plot(range(len(training_ang_errors)),training_ang_errors,'r')
plt.plot(range(len(testing_ang_errors)),testing_ang_errors,'g')
plt.xlabel("Epoch")
plt.ylabel("Mean angular error")
plt.show()
```

As we can see, the model successfully converges with a reasonable amount of generalization error averaging 6% mean angular loss for the validation set. The model appears to have converged without overfitting since the validation error has not started going up, or underfitting since the validation error is no longer continuing to go down at a reasonable rate.

Since the convergence is comfortably below the target error, the model has proven itself to be good enough for submission

```python
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            # state size. 1 x 32 x 32
            nn.Conv2d(1, ndf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf) x 16 x 16
            nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 2),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*2) x 8 x 8
            nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ndf * 4),
            nn.LeakyReLU(0.2, inplace=True),
            # state size. (ndf*4) x 4 x 4
            nn.Conv2d(ndf * 4, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)

netD = Discriminator()
```

```python
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.conv1 = nn.Conv2d(1, ndf, 4, 4, 1, bias=False)
        self.conv2 =nn.Conv2d(ndf, ndf * 2, 4, 2, 1, bias=False)
        self.conv3 =nn.Conv2d(ndf * 2, ndf * 4, 4, 2, 1, bias=False)
        self.conv4 =nn.Conv2d(ndf * 4, 1, 2, 1, 0, bias=False)
        
        self.batch1 =nn.BatchNorm2d(ndf)
        self.batch2 = nn.BatchNorm2d(ndf * 2)
        self.batch3 = nn.BatchNorm2d(ndf * 4)
        
        self.relu =  nn.LeakyReLU(0.2, inplace=True)
        self.sigmoid =  nn.Sigmoid()
        
    def forward(self, x):
        #print(x.shape)
        x = self.conv1(x)
        x = self.batch1(x)
        x = self.relu(x)
        #print(x.shape)

        x = self.conv2(x)
        x = self.batch2(x)
        x = self.relu(x)
        #print(x.shape)
        
        x = self.conv3(x)
        x = self.batch3(x)
        x = self.relu(x)
        #print(x.shape)
        
        x = self.conv4(x)
        #print(x.shape)

        x = self.sigmoid(x)
        #print('final shape')
        #print(x.size())
        #print(x.size())
        return x
        
netD = Discriminator()
```

```python
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.conv_t1 = nn.ConvTranspose2d( nz, ngf * 4, 4, 2, 1,output_padding=(1,0), bias=False)
        self.batch1 = nn.BatchNorm2d(ngf * 4)
        self.relu = nn.ReLU(True)
        self.tanh = nn.Tanh()
        
        self.conv_t2 = nn.ConvTranspose2d(ngf * 4, ngf * 2, 4, 2, 1,output_padding=(1,0), bias=False)
        self.batch2 = nn.BatchNorm2d(ngf * 2)
        
        self.conv_t3 = nn.ConvTranspose2d(in_channels=ngf * 2,out_channels=ngf, 
                                          kernel_size=4, stride=2, padding=1,output_padding=(1,1), 
                                          bias=False)
        self.batch3 = nn.BatchNorm2d(ngf)
       
        self.conv_t4 = nn.ConvTranspose2d( ngf, 1, 4, 2, 1, output_padding=(1,0), bias=False)
        
    def forward(self, x):
        # in: latent_size x 1 x 1
        #print(x.shape)
        x = self.conv_t1(x)
        x = self.batch1(x)
        x = self.relu(x)
        #print(x.shape)
        # out: ngf * 4 x 

        x = self.conv_t2(x)
        #print(x.shape)
        x = self.batch2(x)
        print('conv 2')
        print(x.shape)
        x = self.relu(x)
        # 
        x = self.conv_t3(x)
        x = self.batch3(x)
        x = self.relu(x)
        print('conv 3')
        print(x.shape)
        #print(type(x))
        x = self.conv_t4(x)
        x = self.tanh(x)
        print('conv 4')
        print(x.shape)
        
        return x

netG = Generator()
```

```python
# Initialize BCELoss function
criterion = nn.BCELoss()

# Establish convention for real and fake labels during training
real_label = 1.
fake_label = 0.

# Setup Adam optimizers for both G and D
optimizerD = torch.optim.Adam(netD.parameters(), lr=lr, betas=(beta1, 0.999))
optimizerG = torch.optim.Adam(netG.parameters(), lr=lr, betas=(beta1, 0.999))

# Training Loop

print("Starting Training Loop...")
# For each epoch
for epoch in range(num_epochs):
    # For each batch in the dataloader
    for i, (images,labels) in enumerate(gan_train_loader, 0):

        ############################
        # (1) Update D network: maximize log(D(x)) + log(1 - D(G(z)))
        ###########################
        ## Train with all-real batch
        netD.zero_grad()
        # Format batch
        real_images = images.to(device)
        label = torch.full((batch_size*2,), real_label, dtype=torch.float, device=device)
        # Forward pass real batch through D
        output = netD(real_images).view(-1)
        # Calculate loss on all-real batch
        #print(type(output))
        #print(output.shape)
        #print(label.shape)
        errD_real = criterion(output, label)
        # Calculate gradients for D in backward pass
        errD_real.backward()
        D_x = output.mean().item()

        ## Train with all-fake batch
        # Generate batch of latent vectors
        z = torch.randn(batch_size, nz, 1, 1, device=device)
        # Generate fake image batch with G
        fake = netG(z)
        label.fill_(fake_label)
        # Classify all fake batch with D
        output = netD(fake.detach()).view(-1)
        # Calculate D's loss on the all-fake batch
        print("output size: {}\n label size: {}\n".format(output.shape, label.shape))
        errD_fake = criterion(output, label)
        # Calculate the gradients for this batch, accumulated (summed) with previous gradients
        errD_fake.backward()
        D_G_z1 = output.mean().item()
        # Compute error of D as sum over the fake and the real batches
        errD = errD_real + errD_fake
        # Update D
        optimizerD.step()

        ############################
        # (2) Update G network: maximize log(D(G(z)))
        ###########################
        netG.zero_grad()
        label.fill_(real_label)  # fake labels are real for generator cost
        # Since we just updated D, perform another forward pass of all-fake batch through D
        output = netD(fake).view(-1)
        # Calculate G's loss based on this output
        errG = criterion(output, label)
        # Calculate gradients for G
        errG.backward()
        D_G_z2 = output.mean().item()
        # Update G
        optimizerG.step()

        # Output training stats
        if i % 50 == 0:
            print('[%d/%d][%d/%d]\tLoss_D: %.4f\tLoss_G: %.4f\tD(x): %.4f\tD(G(z)): %.4f / %.4f'
                  % (epoch+1, num_epochs, i, len(dataloader),
                     errD.item(), errG.item(), D_x, D_G_z1, D_G_z2))
```
